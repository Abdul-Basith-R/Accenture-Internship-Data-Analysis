```python
profile["Interests"] = profile["Interests"].apply(eval)
```

```python
interest = {}
for i in profile["Interests"]:
    for j in i:
        if j not in interest:
            interest[j] = 1
        else:
            interest[j] += 1
```

```python
def to_1D(series):
    return pd.Series([x for _list in series for x in _list])
```

```python
to_1D(profile["Interests"]).value_counts()
```

```python
fig, ax = plt.subplots(figsize = (25,10))
ax.bar(to_1D(profile["Interests"]).value_counts().index,
        to_1D(profile["Interests"]).value_counts().values)
ax.set_ylabel("Frequency", size = 20)
ax.set_title("Interests", size = 20)
```
